```python
import geopandas as gp
import numpy as np

import matplotlib.pyplot as plt
# plt.rcParams['figure.figsize'] = (12,6)

%matplotlib inline

PROJ_MERCURY_MERCATOR = '+proj=merc +a=2439700 +b=2439700 +lat_ts=0.0 +lon_0=0.0 +x_0=0.0 +y_0=0 +k=1.0 +units=m +nadgrids=@null +wktext  +no_defs'
```

## Contacts

```python
contacts = gp.read_file('D2.2/Shapefiles/H05_contacts/H05_contacts.shp')
```

```python
contacts.head()
```

```python
contacts.crs
```

```python
contacts_reproj = contacts.to_crs(crs='+proj=merc +a=2439700 +b=2439700 +lat_ts=0.0 +lon_0=0.0 +x_0=0.0 +y_0=0 +k=1.0 +units=m +nadgrids=@null +wktext  +no_defs')
```

```python
len(contacts)
```

```python
fig, ax = plt.subplots(figsize=(14,8))
contacts.plot(ax=ax)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(14,8))
contacts_reproj.plot(ax=ax)
fig.show()
```

```python
contacts_reproj.to_file('D2.2/Shapefiles/H05_contacts/H05_contacts-MERCATOR.shp')
```

## Geological units

```python
geounits = gp.read_file('D2.2/Shapefiles/H05_geological_units/3_classes/H05_geological_units_3_classes.shp')
geounits.crs
```

```python
geounits.head()
```

```python
fig, ax = plt.subplots(figsize=(14,8))
geounits.plot(ax=ax)
fig.show()
```

```python
geounits['Geological'].unique()
```

```python
fig, ax = plt.subplots(figsize=(14,8))
geounits.plot(ax=ax, column='Geological', cmap='tab10');
```

```python
geounits_reproj = geounits.copy()
geounits_reproj.to_crs(PROJ_MERCURY_MERCATOR, inplace=True)
geounits_reproj.crs
```

```python
fig, ax = plt.subplots(figsize=(14,8))
geounits_reproj.plot(ax=ax, column='Geological', cmap='tab10')
```

```python
geounits_reproj.to_file('D2.2/Shapefiles/H05_geological_units/3_classes/H05_geological_units_3_classes-MERCATOR.shp')
```
